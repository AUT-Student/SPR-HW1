# Load Dataset

```python
import pandas as pd
data = pd.read_csv("iris.data", header=None, names=["Sepal Length", "Sepal Width", "Petal Length", "Petal Width", "class"])
```

# a

```python
class_names = list(set(data["class"].values))
```

```python
class_names
```

```python
# Seperate Data Class
data_class = []
for i, class_name in enumerate(class_names):
    data_class.append(data[data["class"] == class_name])
```

```python
import matplotlib.pyplot as plt

class_colors = ["black", "red", "blue"]

fig, axs = plt.subplots(1,3)

# Feature 1 and 2
for i in range(len(data_class)):
    axs[0].scatter(data_class[i]["Sepal Length"].values, data_class[i]["Sepal Width"].values,
                color=class_colors[i], label=class_names[i])


axs[0].title.set_text("Distribution of Feature 1 and 2\n(Sepal Length and Sepal Width)")    
axs[0].legend()
plt.setp(axs[0], xlabel='Sepal Length (cm)')
plt.setp(axs[0], ylabel='Sepal Width (cm)')    

# Feature 1 and 3
for i in range(len(data_class)):
    axs[1].scatter(data_class[i]["Sepal Length"].values, data_class[i]["Petal Length"].values,
                color=class_colors[i], label=class_names[i])


axs[1].title.set_text("Distribution of Feature 1 and 3\n(Sepal Length and Petal Length)")    
axs[1].legend()
plt.setp(axs[1], xlabel='Sepal Length (cm)')
plt.setp(axs[1], ylabel='Petal Length (cm)')    


# Feature 2 and 4
for i in range(len(data_class)):
    axs[2].scatter(data_class[i]["Sepal Width"].values, data_class[i]["Petal Width"].values,
                color=class_colors[i], label=class_names[i])


axs[2].title.set_text("Distribution of Feature 2 and 4\n(Sepal Width and Petal Width)")    
axs[2].legend()
plt.setp(axs[2], xlabel='Sepal Width (cm)')
plt.setp(axs[2], ylabel='Petal Width (cm)')    

# Plot all!
fig.set_size_inches(17,5)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

class_colors = ["black", "red", "blue"]

fig, axs = plt.subplots(4,3)

for i, t in enumerate([ [1.2, -0.3], [-1.8, 0.6], [1.4, 0.5], [-0.5, -1]]):
    # Feature 1 and 2
    for j in range(len(data_class)):
        axs[i][0].scatter(data_class[j]["Sepal Length"].values * t[0] + data_class[j]["Sepal Width"].values * t[1],
                          np.full(len(data_class[j]), j), color=class_colors[j], label=class_names[j], alpha=0.2)


    axs[i][0].title.set_text(f"Linear Transformation T{i+1} on Distribution of Feature 1 and 2\n(Sepal Length and Sepal Width)")    
    axs[i][0].legend()
    plt.setp(axs[i][0], xlabel=f"T{i+1} Value (cm)", ylabel="Class")

    # Feature 1 and 3
    for j in range(len(data_class)):
        axs[i][1].scatter(data_class[j]["Sepal Length"].values * t[0] + data_class[j]["Petal Length"].values * t[1],
                          np.full(len(data_class[j]), j), color=class_colors[j], label=class_names[j], alpha=0.2)


    axs[i][1].title.set_text(f"Linear Transformation T{i+1} on Distribution of Feature 1 and 3\n(Sepal Length and Petal Length)")    
    axs[i][1].legend()
    plt.setp(axs[i][1], xlabel=f"T{i+1} Value (cm)", ylabel="Class")

    # Feature 2 and 4
    for j in range(len(data_class)):
        axs[i][2].scatter(data_class[j]["Sepal Width"].values * t[0] + data_class[j]["Petal Width"].values * t[1],
                          np.full(len(data_class[j]), j), color=class_colors[j], label=class_names[j], alpha=0.2)


    axs[i][2].title.set_text(f"Linear Transformation T{i+1} on Distribution of Feature 2 and 4\n(Sepal Width and Petal Width)")    
    axs[i][2].legend()
    plt.setp(axs[i][2], xlabel=f"T{i+1} Value (cm)", ylabel="Class")

# Plot all!
fig.set_size_inches(17,12)
fig.tight_layout()
plt.show()
```
